```python
x = df.drop(columns = ['price'], axis = 1)
y = df['price']
```

```python
x
```

```python
y
```

```python
df.columns
```

```python
x_one_encode = pd.get_dummies(x,columns = ['model','transmission','fuelType'],drop_first = True)
```

```python
x_one_encode
```

```python
x_one_encode.columns
```

```python
x_one_encode = x_one_encode.astype(int)
```

```python
x_one_encode
```

```python
from sklearn.preprocessing import LabelEncoder

encode = LabelEncoder()
```

```python
columns = ['model','transmission','fuelType']
xlable = x

for i in columns:
    xlable[i] = encode.fit_transform(xlable[i])
```

```python
xlable
```

```python
df['model'].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
numerical_col = ['year','mileage', 'tax', 'mpg' , 'engineSize']
```

```python
scaler = StandardScaler()
x_one_encode[numerical_col] = scaler.fit_transform(x_one_encode[numerical_col])
```

```python
x_one_encode
```

```python
xlable.columns
```

```python
xlable[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']] = scaler.fit_transform(xlable[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']])
```

```python
xlable
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_one_encode, y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
r2 = r2_score(y_test,y_pred)
r2
```

```python
n = X_test.shape[0]
p = X_test.shape[1]
adjusted_r2 = 1 - ((1 - r2) * (n - 1)) / (n - p - 1)
print("Adjusted R² Score:", adjusted_r2)
```
